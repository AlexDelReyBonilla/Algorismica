### 2.9 Min max

Donada una llista de nombres enters, troba el valor mínim i el valor màxim amb un algorisme de complexitat $O(n)$. No utilitzis cap funció específica de Python (com ara les funcions min(), max() o sort()) per resoldre el problema.

```python
def minim_maxim(llista):
    min = llista[0]
    max = llista[0]
    for i in range(1, len(llista)):
        if llista[i] < min:
            min = llista[i]
        if llista[i]> max:
            max = llista[i]
    """
    Aquesta funció troba el valor mínim i el valor màxim.
    
    Parameters
    ----------
    llista: list
        La llista d'enters
    
    Returns
    -------
    min, max: int, int
        Els valors mínim i màxim dins la llista
    """
   
    return min, max
minim_maxim ([1, 7, 4, 6, 8, -2, 9, 5])
```

```python
assert minim_maxim([3, 1, 5, 2, 7, 8]) == (1, 8)
assert minim_maxim([1, 7, 4, 6, 8, -2, 9, 5]) == (-2, 9)
```
